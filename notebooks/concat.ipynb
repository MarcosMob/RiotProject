```python
import glob
import pandas as pd
```

```python
def concat_csv(archives_path):
#Coletanado todos os arquivos com .csv da pasta Csv
    archives_path = archives_path + "/*.csv"

    csv_archives = glob.glob(archives_path)
    dfs = []

    for archive in csv_archives:
        df = pd.read_csv(archive)
        dfs.append(df)

    df_final = pd.concat(dfs, ignore_index = True)
    return df_final
```

```python
path = '/Users/usuario/Documentos/riot_project_ds/learning/Arthur_lanches_games_CSVs'
df_final = concat_csv(path)
df_final
```
